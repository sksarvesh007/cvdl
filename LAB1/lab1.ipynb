## TASK1 : FLIPPING THE IMAGE USING THE NUMPY ARRAYS

```python
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
plt.imshow(gray, cmap= 'gray')
grayf = gray.copy()
sft = gray.copy()
```

```python
row,col = gray.shape
for r in range(row):
    for c in range(int(col/2)):
        temp = grayf[r][c]
        grayf[r][c] = grayf[r][col-c-1]
        grayf[r][col-c-1] = temp
```

```python
plt.imshow(grayf, cmap='gray')
```

```python
row,col = gray.shape
sft = gray.copy()
for r in range(row):
    for c in range(col-51):
        sft[r][c] = sft[r][c+40]
```

```python
plt.imshow(sft, cmap='gray')
```

```python
rect = gray.copy()
cv2.rectangle(rect, (50,100),(470,510),(0,0,0),3)
plt.imshow(rect)
```

## TASK 2 : BLUR THE IMAGE USING GAUSSIAN AND SIMPLE BLULR FILTER 

```python
import numpy as np
average= gray.copy()
average = average.astype(np.float64)
r , c = average.shape
avg_upd = average.copy()
for i in range(1,r-1):
    for j in range(1,c-1):
        avg_upd[i][j] = (average[i-1][j-1] + average[i-1][j] + average[i-1][j+1] + average[i][j-1] + average[i][j] + average[i][j+1] + average[i+1][j-1] + average[i+1][j] + average[i+1][j+1])/9
```

```python
#show blurred and non blurred image side by side
fig, ax = plt.subplots(1,2)
ax[0].imshow(gray, cmap='gray')
ax[1].imshow(avg_upd, cmap='gray')
#give title as well 
ax[0].set_title('Original Image')
ax[1].set_title('Blurred Image')
```

```python
simple_blur = cv2.blur(gray, (5,5))
gauss_blur = cv2.GaussianBlur(gray, (5,5), 0)
fig , ax = plt.subplots(1,3)
ax[0].imshow(simple_blur, cmap='gray')
ax[1].imshow(gauss_blur, cmap='gray')
ax[2].imshow(gray, cmap = 'gray')
ax[0].set_title('Simple Blur')
ax[1].set_title('Gaussian Blur')
ax[2].set_title('Original Image')
```
