```python
import cv2 
import matplotlib.pyplot as plt 
```

```python
image = cv2.imread('C:/Users/Sarvesh/Desktop/repos/cvdl/image.jpeg')
```

```python
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image)
```

```python

sobel_x = cv2.Sobel(image, cv2.CV_64F, 1, 0, ksize=5)
sobel_y = cv2.Sobel(image, cv2.CV_64F, 0, 1, ksize=5)

sobel_x = cv2.convertScaleAbs(sobel_x)

sobel_y = cv2.convertScaleAbs(sobel_y)

sobel = cv2.addWeighted(sobel_x, 0.5, sobel_y, 0.5, 0)
plt.figure(figsize=(20,20))
plt.subplot(1,3,1)
plt.imshow(sobel_x)
plt.title('Sobel X, Vertical Edges')
plt.subplot(1,3,2)
plt.imshow(sobel_y)
plt.title('Sobel Y, Horizontal Edges')
plt.subplot(1,3,3)
plt.imshow(sobel)
plt.title('Sobel, Combined Edges')
plt.show()
```

```python
grey_scale = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
canny = cv2.Canny(grey_scale, 100, 400)
plt.subplot(1,2,1)
plt.imshow(grey_scale, cmap='gray')
plt.title('Grey Scale Image')
plt.subplot(1,2,2)
plt.imshow(canny,  cmap='gray')
plt.title('Canny Edge Detection')
```

```python
log = cv2.Laplacian(grey_scale, cv2.CV_64F)
log = cv2.convertScaleAbs(log)
plt.figure(figsize=(8,8))
plt.subplot(1,2,1)
plt.imshow(grey_scale, cmap='gray')
plt.title('Grey Scale Image')
plt.subplot(1,2,2)
plt.imshow(log, cmap='gray')
plt.title('Log Edge Detection')
plt.show()
```

```python
import numpy as np
image = cv2.imread('C:/Users/Sarvesh/Desktop/repos/cvdl/image.jpeg')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
gray = np.float32(gray)
dst = cv2.cornerHarris(gray, 2, 3, 0.04)
dst = cv2.dilate(dst, None)
image[dst>0.01*dst.max()] = [255, 0, 0]
plt.imshow(image)
plt.title('Harris Corner Detection')
plt.show()
```
